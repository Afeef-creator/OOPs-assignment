**Theory Questions**

1. What is Object-Oriented Programming (OOP)?
    - It's basically a programming style where we organize our code around objects instead of just functions and logic. Think of it like modeling real-world stuffâ€”like a car or a dogâ€”with code. It makes the code way more organized, reusable, and easier to manage, especially for big projects.

2. What is a class in OOP?
    - A class is like a blueprint or a template for creating objects. It defines what the objects of this type will have (data/attributes) and what they will be able to do (methods/functions). For example, a "Dog" class defines that all dogs have a name and breed (attributes) and can bark and run (methods).

3. What is an object in OOP?
    - An object is a specific instance of a class. It's the actual thing created from the blueprint. So, if "Dog" is the class (blueprint), then my specific dog, "Buddy", with a name "Buddy" and breed "Golden Retriever," is an object. They have real values for the attributes defined in the class.

4. What is the difference between abstraction and encapsulation?
    - Abstraction is about showing only the necessary details and hiding the complex implementation. It's like driving a car: you only need to know how to use the steering wheel and pedals, not how the engine works internally.

    - Encapsulation is about bundling data (attributes) and the methods that operate on that data together inside a class. It also involves restricting direct access to some components, like putting private stuff in a protective capsule.

    - They work together, but Abstraction is about what to show, and Encapsulation is about how to protect and bundle the data.

5. What are dunder methods in Python?
    - Dunder methods (short for Double Underscore) are special methods in Python that have two underscores before and after their name, like __init__ or __str__. They are not meant to be called directly by us, but are hooks that Python calls automatically in specific situations. For instance, __init__ is called when you create an object, and __str__ is called when you try to print an object.

6. Explain the concept of inheritance in OOP?
    - Inheritance is when a new class (the child or subclass) derives properties and behavior from an existing class (the parent or superclass). It's like real-world inheritance: a "Puppy" class can inherit all the attributes and methods from a "Dog" class, so you don't have to rewrite the "bark" or "run" methods again. It promotes code reusability.

7. What is polymorphism in OOP?
    - Polymorphism literally means "many forms." In OOP, it means that an object can take on many forms or, more practically, that methods can do different things depending on the object they are called on. A classic example is a speak() method: if called on a Dog object, it barks; if called on a Cat object, it meows. The same method name acts differently based on the object's type.

8. How is encapsulation achieved in Python?
    - Python doesn't have strict "private" keywords like some other languages. Encapsulation is achieved mostly by convention.

        - We use a single leading underscore (e.g., _name) to indicate that an attribute is "protected" and should be treated as private, meaning external code shouldn't mess with it directly.

        - We use a double leading underscore (e.g., __name) for name mangling, which Python changes internally to make it harder, though not impossible, for subclasses to accidentally overwrite it.

9. What is a constructor in Python?
    - A constructor is a special method in a class that gets called automatically when a new object (instance) is created. In Python, this method is the __init__ dunder method. Its main job is to initialize the object's attributes (set their initial values) and get the object ready for use.

10. What are class and static methods in Python?

    - Instance Method
      - Normal function in class.
      - it takes as first argument	self (the instance itself)
      - it can access Instance data and class data
      - Use Case Operating on a specific object's data.
    - Class Method
      - Decorated with @classmethod
      - it takes as first argument cls (the class itself)
      - it can access Class data, but not instance data.
      - Use Case Factory methods, or methods that operate on the class state.
    - Static Method
      - Decorated with @staticmethod
      - it takes as first argument
      - it can access No special first argument	Nothing from the class or instance.
      - Use Case Utility functions that logically belong to the class but don't need its data.

11. What is method overloading in Python?
    - Method overloading means having multiple methods with the same name in a single class, but with different numbers or types of arguments.

    - Python doesn't support true method overloading like Java or C++. If you define two methods with the same name, the last one defined simply overwrites the earlier ones. To achieve similar functionality, Python programmers usually use default arguments or check the type/number of arguments inside a single method.

12. What is method overriding in OOP?
    - Method overriding is when a subclass provides a new implementation for a method that is already defined in its superclass. This is crucial for inheritance and polymorphism. For example, if the parent Animal class has a generic speak() method, the child Dog class can override it to have a specific bark() sound.

13. What is a property decorator in Python?
    - The @property decorator is a way to give your attributes getter, setter, and deleter functionality without having to call them as methods. It lets you write code that looks like you are directly accessing or setting an attribute (e.g., obj.name), but under the hood, it calls a method to perform checks, validations, or calculations. It's great for encapsulation!

14. Why is polymorphism important in OOP?
    - Polymorphism is super important because it leads to flexible and decoupled code.

        - Flexibility: You can write code that works with a general type of object (like an Animal) and it will automatically handle the specific behavior of any subclass (like Dog or Cat) without needing lots of if/else checks for every type.

        - Decoupling: Your code that uses the objects doesn't need to know the specific type of object it's dealing with, only that it has a method with a certain name.

15. What is an abstract class in Python?
    - An abstract class is a class that cannot be instantiated (you can't create an object directly from it). It's meant to be a blueprint for other classes to inherit from. Abstract classes often contain abstract methods (methods that are declared but have no implementation), forcing the subclasses to provide their own specific implementations. You use the abc (Abstract Base Class) module in Python to create them.

16. What are the advantages of OOP?
    - The main perks we get from using OOP are:

        - Code Reusability: Through inheritance, we don't have to write the same code multiple times.

        - Maintainability: The code is better organized (encapsulation), so finding and fixing bugs is easier.

        - Flexibility (Polymorphism): We can write general, adaptable code.

        - Better Modeling: It helps us model complex, real-world problems in a more intuitive way.
17. What is the difference between a class variable and an instance variable?
    - Instance Variable
      - Scope	Unique to each object (instance).
      - Definition	Defined inside methods (usually __init__) using self..
      - Accessed via self.variable_name or instance.variable_name.
      - Example	self.name (My dog's name is "Buddy").
    - Class Variable
      - Scope Shared by all objects of the class.
      - Defined directly inside the class, but outside of any method.
      - Accessed via class_name.variable_name or instance.variable_name.
      - species = "Canis familiaris" (All dogs are this species)

18. What is multiple inheritance in Python?
    - Multiple inheritance is when a child class inherits from more than one parent class. For example, a FlyingDog class could inherit from both an Animal class and an Aircraft class. Python does support it, but it can sometimes lead to complexity (like the "Diamond Problem"), so people usually try to use it carefully or prefer composition.

19. Explain the purpose of __str__ and __repr__ methods in Python.
    - Both are dunder methods used to provide a string representation of an object, but they have different goals:

        - __str__: Meant for the end-user or for human-readable output. It's what Python calls when you use print(object). It should be concise and easily understandable.

        - __repr__: Meant for the developer. It should be an unambiguous string that, ideally, could be used to recreate the object (like a copy-pasteable constructor call). It's what Python calls when you just type the object's name in the console.

20. What is the significance of the super() function in Python?
    - The super() function is a built-in helper that allows a subclass to call methods (like __init__ or any other method) that are defined in its parent class. Its main significance is to make sure the parent class's initialization or logic is executed when the child class is running its own. It's essential for properly implementing inheritance, especially with __init__.

21. What is the significance of the __del__ method in Python?
    - The __del__ dunder method is the destructor. It's called by Python's garbage collector when an object is about to be destroyed and its memory is reclaimed. It's rarely used, but when it is, it's typically to perform cleanup actions like closing a file connection or releasing an external resource associated with the object before it vanishes.
  
22. What is the difference between @staticmethod and @classmethod in Python?
    - @staticmethod
      - First Argument	None (It's just a regular function).
      - it sees	Nothing about the class or the instance.
      - Use Case	Utility function that has a logical connection to the class but doesn't need its internal state.
    - @classmethod
      - First Argument cls (refers to the class itself)
      - it sees The class itself and its variables.
      - Use Case Factory methods, or methods that need to access or modify class-level state (like class variables).
    
23. How does polymorphism work in Python with inheritance?
    - Polymorphism works through method overriding. If a superclass and a subclass both have a method with the same name (e.g., drive()), Python automatically calls the correct version of the method based on the type of the object that the method is being called on. If you have a list of different Vehicle objects and loop through them calling vehicle.drive(), the Car object will run its drive() method, and the Truck object will run its (overridden) drive() method.

24. What is method chaining in Python OOP?
    - Method chaining is a technique where you call multiple methods on the same object sequentially in a single line. This works only if each method in the chain returns self (the instance itself) after it finishes its job. It makes the code very readable and concise, often used for configuring objects, like: my_car.start_engine().set_gear(1).accelerate(50)

25. What is the purpose of the __call__ method in Python?
    - The __call__ dunder method makes an object callable as if it were a function. If you define this method in a class, you can create an object and then call that object directly using parentheses, like my_object(arg1, arg2). This is super useful for making objects act like functions, like when creating custom decorators or stateful function-like objects.

```python
# 1. Create a parent class Animal with a method speak() that prints a generic message.
# Create a child class Dog that overrides the speak() method to print "Bark!".
class Animal:
    """Parent class with a generic speak method."""
    def speak(self):
        print("A generic animal sound.")

class Dog(Animal):
    """Child class overriding the speak method."""
    def speak(self):
        print("Bark! ðŸ•")

# Demonstration
print("--- 1. Method Overriding ---")
Dog().speak()
```

```python
# 2. Write a program to create an abstract class Shape with a method area().
# Derive classes Circle and Rectangle from it and implement the area() method in both.
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    """Abstract Base Class (Blueprint) for shapes."""
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    """Derived class implementing the area() method for a circle."""
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

class Rectangle(Shape):
    """Derived class implementing the area() method for a rectangle."""
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Demonstration
print("\n--- 2. Abstract Class (Shape) ---")
my_circle = Circle(5)
print(f"Circle Area: {my_circle.area():.2f}")
```

```python
# 3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type.
# Derive a class Car and further derive a class ElectricCar that adds a battery attribute.
class Vehicle:
    """Base Class: General vehicle."""
    def __init__(self, vehicle_type):
        self.type = vehicle_type

class Car(Vehicle):
    """Intermediate Class: Inherits from Vehicle."""
    def __init__(self, make, model):
        super().__init__("Car")
        self.make = make
        self.model = model

class ElectricCar(Car):
    """Derived Class: Inherits from Car (multi-level)."""
    def __init__(self, make, model, battery_kwh):
        super().__init__(make, model)
        self.battery = f"{battery_kwh} kWh"

# Demonstration
print("\n--- 3. Multi-level Inheritance ---")
my_tesla = ElectricCar("Tesla", "Model Y", 75)
print(f"{my_tesla.make} is a {my_tesla.type} with a {my_tesla.battery} battery.")
```

```python
# 4. Demonstrate polymorphism by creating a base class Bird with a method fly().
# Create two derived classes Sparrow and Penguin that override the fly() method.
class Bird:
    """Base class with the common method fly()."""
    def fly(self):
        print("Bird is flying high in the sky.")

class Sparrow(Bird):
    """Overrides fly() for a typical flying bird."""
    def fly(self):
        print("Sparrow is fluttering quickly.")

class Penguin(Bird):
    """Overrides fly() to reflect its unique behavior."""
    def fly(self):
        print("Penguin can't fly, but it is swimming fast! ðŸ§")

# Demonstration: Polymorphism in action
print("\n--- 4. Polymorphism (Bird) ---")
birds = [Sparrow(), Penguin()]
for bird in birds:
    bird.fly()
```

```python
# 5. Write a program to demonstrate encapsulation by creating a class BankAccount with private
# attributes balance and methods to deposit, withdraw, and check balance.
class BankAccount:
    """Demonstrates encapsulation with a private __balance attribute."""
    def __init__(self, initial_amount):
        # Private attribute (name mangled)
        self.__balance = initial_amount

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}.")
        else:
            print("Withdrawal failed: Invalid amount or insufficient funds.")

    # Public method (getter) to safely access the attribute
    def check_balance(self):
        return f"Current balance: ${self.__balance}"

# Demonstration
print("\n--- 5. Encapsulation (BankAccount) ---")
account = BankAccount(1000)
account.deposit(500)
account.withdraw(200)
print(account.check_balance())
```

```python
# 6. Demonstrate runtime polymorphism using a method play() in a base class Instrument.
# Derive classes Guitar and Piano that implement their own version of play().
class Instrument:
    """Base class defining a common interface: play()."""
    def play(self):
        print("An instrument is making a sound.")

class Guitar(Instrument):
    """Specific implementation of play() for Guitar."""
    def play(self):
        print("ðŸŽ¸ Strumming a chord on the Guitar.")

class Piano(Instrument):
    """Specific implementation of play() for Piano."""
    def play(self):
        print("ðŸŽ¹ Tapping the keys of the Piano.")

# Demonstration
def play_music(instrument):
    instrument.play()

print("\n--- 6. Runtime Polymorphism (Instrument) ---")
play_music(Guitar())
play_music(Piano())
```

```python
# 7. Create a class MathOperations with a class method add_numbers() to add two numbers
# and a static method subtract_numbers() to subtract two numbers.
class MathOperations:
    """Class demonstrating class methods and static methods."""

    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b

# Demonstration: Call directly using the Class Name
print("\n--- 7. Class and Static Methods ---")
sum_result = MathOperations.add_numbers(20, 15)
diff_result = MathOperations.subtract_numbers(50, 25)
print(f"Sum: {sum_result}, Difference: {diff_result}")
```

```python
# 8. Implement a class Person with a class method to count the total number of persons created.
class Person:
    """Uses a class variable and a class method to track instances."""
    # Class variable, shared by all instances
    total_persons = 0

    def __init__(self, name):
        self.name = name
        # Increment the shared variable when a new instance is created
        Person.total_persons += 1

    @classmethod
    def count_persons(cls):
        """Returns the total number of Person objects created."""
        return cls.total_persons

# Demonstration
print("\n--- 8. Instance Counting with Class Method ---")
p1 = Person("Alice")
p2 = Person("Bob")
print(f"Total persons created: {Person.count_persons()}")
```

```python
# 9. Write a class Fraction with attributes numerator and denominator.
# Override the str method to display the fraction as "numerator/denominator".
class Fraction:
    """Overrides __str__ to provide a user-friendly string representation."""
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        """Defines the string returned when print() or str() is called."""
        return f"{self.numerator}/{self.denominator}"

# Demonstration
print("\n--- 9. Overriding __str__ (Fraction) ---")
f = Fraction(3, 4)
print(f"The fraction is: {f}")
```

```python
# 10. Demonstrate operator overloading by creating a class Vector and
# overriding the add method to add two vectors.
class Vector:
    """Overloads the + operator using the __add__ dunder method."""
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

    def __add__(self, other):
        """Implements vector addition for the '+' operator."""
        if isinstance(other, Vector):
            new_x = self.x + other.x
            new_y = self.y + other.y
            return Vector(new_x, new_y)
        else:
            raise TypeError("Can only add a Vector to another Vector.")

# Demonstration
print("\n--- 10. Operator Overloading (Vector Addition) ---")
v1 = Vector(2, 5)
v2 = Vector(10, 1)
v3 = v1 + v2 # Calls v1.__add__(v2)
print(f"Result (v1 + v2): {v3}")
```

```python
# 12. Implement a class Student with attributes name and grades.
# Create a method average_grade() to compute the average of the grades.
class Student:
    """Class to manage student grades and calculate the average."""
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        """Computes the average of the student's grades."""
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)

# Demonstration
print("\n--- 12. Student Average Grade ---")
student_a = Student("Maya", [90, 85, 95, 88])
avg = student_a.average_grade()
print(f"{student_a.name}'s average grade is: {avg:.2f}")
```

```python
# 13. Create a class Rectangle with methods set_dimensions() to set the dimensions
# and area() to calculate the area.
class Rectangle:
    """Class to represent a rectangle."""
    def __init__(self):
        self.length = 0
        self.width = 0

    def set_dimensions(self, length, width):
        """Method to set the dimensions of the rectangle."""
        self.length = length
        self.width = width

    def area(self):
        """Calculates and returns the area."""
        return self.length * self.width

# Demonstration
print("\n--- 13. Rectangle Dimensions and Area ---")
rect = Rectangle()
rect.set_dimensions(12, 6)
print(f"Rectangle Area: {rect.area()}")
```

```python
# 14. Create a class Employee with a method calculate_salary() that computes the salary
# based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.
class Employee:
    """Base class for all employees with base salary calculation."""
    def __init__(self, hours_worked, hourly_rate):
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

class Manager(Employee):
    """Derived class that adds a bonus to the base salary."""
    def __init__(self, hours_worked, hourly_rate, bonus):
        super().__init__(hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        """Overrides and extends the base salary calculation."""
        base_salary = super().calculate_salary()
        return base_salary + self.bonus

# Demonstration
print("\n--- 14. Inheritance and Salary Extension ---")
mgr = Manager(160, 30, 1000)
print(f"Manager Salary: ${mgr.calculate_salary():.2f}")
```

```python
# 15. Create a class Product with attributes name, price, and quantity.
# Implement a method total_price() that calculates the total price of the product.
class Product:
    """Class to manage product details and calculate total price."""
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        """Calculates the total cost of the product quantity."""
        return self.price * self.quantity

# Demonstration
print("\n--- 15. Product Total Price ---")
item = Product("Laptop", 1200.50, 2)
print(f"Total Price for {item.name}: ${item.total_price():.2f}")
```

```python
# 16. Create a class Animal with an abstract method sound().
# Create two derived classes Cow and Sheep that implement the sound() method.
from abc import ABC, abstractmethod

class Animal(ABC):
    """Abstract class enforcing a 'sound' method."""
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    """Implements the abstract sound() method."""
    def sound(self):
        return "Moo!"

class Sheep(Animal):
    """Implements the abstract sound() method."""
    def sound(self):
        return "Baa!"

# Demonstration
print("\n--- 16. Abstract Animal Sound ---")
print(f"Cow says: {Cow().sound()}")
```

```python
# 17. Create a class Book with attributes title, author, and year_published.
# Add a method get_book_info() that returns a formatted string with the book's details.
class Book:
    """Class to hold book information and format its details."""
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        """Returns a nicely formatted string of the book's details."""
        return (
            f"Title: {self.title} | "
            f"Author: {self.author} | "
            f"Published: {self.year_published}"
        )

# Demonstration
print("\n--- 17. Book Information ---")
book1 = Book("Python OOP", "Student X", 2025)
print(book1.get_book_info())
```

```python
# 18. Create a class House with attributes address and price.
# Create a derived class Mansion that adds an attribute number_of_rooms.
class House:
    """Base class for residential property."""
    def __init__(self, address, price):
        self.address = address
        self.price = price

class Mansion(House):
    """Derived class that adds a number_of_rooms attribute."""
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

    def get_details(self):
        return (
            f"Address: {self.address}, Price: ${self.price:,.0f}, "
            f"Rooms: {self.number_of_rooms}"
        )

# Demonstration
print("\n--- 18. House and Mansion Inheritance ---")
mansion = Mansion("123 Grand Ave", 15000000, 25)
print(mansion.get_details())
```
